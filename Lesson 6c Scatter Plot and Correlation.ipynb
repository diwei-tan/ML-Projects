# Lesson 6c Scatter Plot and Correlation

Examples of scatter plot and correlation as one of the useful data exploration tools


```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

```python
# importing the Iris dataset with pandas
dataset = pd.read_csv('Iris.csv')
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(dataset.iloc[:,1:5], alpha=0.2, figsize=(6, 6), diagonal='kde')
```

```python
import seaborn as sns
sns.set(style="ticks")

sns.pairplot(dataset, hue="Species")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

cax = ax1.imshow(dataset.corr(), interpolation="nearest")
ax1.grid(True)
plt.title('Correlation ')
labels=['Sex','Length','Diam','Height','Whole','Shucked','Viscera','Shell','Rings',]
ax1.set_xticklabels(labels,fontsize=6)
ax1.set_yticklabels(labels,fontsize=6)
# Add colorbar, make sure to specify tick locations to match desired ticklabels
fig.colorbar(cax, ticks=[-0.4,0,.4,.6,.8,.9,1])
plt.show()
```
